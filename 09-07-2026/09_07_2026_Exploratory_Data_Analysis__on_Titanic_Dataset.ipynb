# Titanic Dataset Exploratory Data Analysis (EDA) Report

## 1. Introduction

The objective of this project is to perform Exploratory Data Analysis (EDA) on the Titanic dataset. The analysis aims to understand passenger demographics, travel patterns, fare distributions, family relationships, and survival trends using statistical methods and visualizations.

The project includes:

* Data Understanding

## 2. Dataset Overview

The dataset contains information about Titanic passengers, including:

* **PassengerId**: Unique identifier for each passenger.
* **Survived**: Survival status (0 = No, 1 = Yes).
* **Pclass**: Passenger class (1 = 1st, 2 = 2nd, 3 = 3rd).
* **Name**: Passenger's name.
* **Sex**: Gender of the passenger.
* **Age**: Age in years.
* **SibSp**: Number of siblings/spouses aboard.
* **Parch**: Number of parents/children aboard.
* **Ticket**: Ticket number.
* **Fare**: Passenger fare.
* **Cabin**: Cabin number.
* **Embarked**: Port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton).

### Dataset Dimensions

The dataset consists of passenger-related information used for analysis of travel characteristics and survival patterns.

## 3. Data Cleaning

### Missing Values

The dataset contained missing values in the following columns:

* **Age**
* **Fare**
* **Cabin**

### Handling Missing Values

*   **Age**: Missing values in the Age column were replaced using the **median** value because the median is less affected by outliers.
*   **Fare**: The missing value in the Fare column was replaced with the **median fare**.
*   **Cabin**: Missing values in the Cabin column were replaced with "**Unknown**" to preserve the column and retain useful information.

### Duplicate Records

The dataset was checked for duplicate records, and no significant duplicate entries were found.
