```python
#stratified sampling
import random

# Population divided into strata (departments)
population = {
"HR": ["EMP101", "EMP102", "EMP103", "EMP104"],
"IT": ["EMP201", "EMP202", "EMP203", "EMP204", "EMP205"],
"Finance": ["EMP301", "EMP302", "EMP303"]
}

# Randomly select samples from each department
hr_sample = random. sample(population["HR"], 2)
it_sample = random.sample(population["IT"], 2)
finance_sample = random.sample(population["Finance"], 1)

# Combine all selected samples
stratified_sample = hr_sample + it_sample + finance_sample

# Display results
print("Population:")
for department, employees in population.items():
  print(f"{department}: {employees}")

print("\nStratified Sample:")
print(stratified_sample)
```

```python
import random

# Population divided into clusters (branches)
population = {
    "Bangalore Branch": ["EMP101", "EMP102", "EMP103", "EMP104"],
    "Hyderabad Branch": ["EMP201", "EMP202", "EMP203", "EMP204"],
    "Chennai Branch": ["EMP301", "EMP302", "EMP303", "EMP304"],
    "Mumbai Branch": ["EMP401", "EMP402", "EMP403", "EMP404"]
}

# Set seed for reproducibility
random.seed(42)

# Select one cluster randomly
selected_cluster = random.choice(list(population.keys()))

# Take all members from the selected cluster
cluster_sample = population[selected_cluster]

# Display population
print("Population:")
for branch, employees in population.items():
    print(f"{branch}: {employees}")

# Display selected cluster and sample
print("\nSelected Cluster:", selected_cluster)
print("Cluster Sample:", cluster_sample)
```

```python
from scipy.stats import ttest_ind

office = df[df["Work_Mode"]=="Office"]["Monthly_Watch_Time"]

hybrid = df[df["Work_Mode"]=="Hybrid"]["Monthly_Watch_Time"]

print("Office Mean =", office.mean())
print("Hybrid Mean =", hybrid.mean())

t_stat, p_value = ttest_ind(office, hybrid)

print("T-statistic =", t_stat)
print("P-value =", p_value)

if p_value < 0.05:
    print("Significant Difference")
else:
    print("No Significant Difference")
```

```python
from scipy.stats import sem, t

confidence = 0.95

salary = df["Salary"]

n = len(salary)

mean = salary.mean()

se = sem(salary)

margin = t.ppf((1+confidence)/2, n-1) * se

lower = mean - margin
upper = mean + margin

print("95% Confidence Interval:")
print(lower, upper)
```

```python
ai = df[df["Department"]=="AI"]["Performance_Rating"]

finance = df[df["Department"]=="Finance"]["Performance_Rating"]

t_stat, p = ttest_ind(ai, finance)

print("T-statistic =", t_stat)
print("P-value =", p)

if p < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

```python
from scipy.stats import f_oneway

office = df[df["Work_Mode"]=="Office"]["Monthly_Sales"]

hybrid = df[df["Work_Mode"]=="Hybrid"]["Monthly_Sales"]

remote = df[df["Work_Mode"]=="Remote"]["Monthly_Sales"]

F, p = f_oneway(office, hybrid, remote)

print("F =", F)
print("P-value =", p)

if p < 0.05:
    print("Monthly Sales differ significantly")
else:
    print("No significant difference")
```
