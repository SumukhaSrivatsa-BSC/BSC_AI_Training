```python
#Building the CNN Model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

```python
# Create a Sequential model (layers are added one after another)
model = Sequential([

# First Convolution Layer
# 32 filters (feature detectors)
# (3,3) filter/kernel size
# ReLU activation introduces non-linearity
# Input image size = 224 x 224 with 1 channel (grayscale)
Conv2D(32, (3,3), activation='relu',
input_shape=(224,224,1)),

# First Max Pooling Layer
# Uses a 2x2 window
# Reduces image dimensions by taking the maximum value
# Helps reduce computation and prevents overfitting
MaxPooling2D((2,2)),

# Second Convolution Layer
# 64 filters are used to learn more complex features
# such as shapes, corners, textures, etc.
Conv2D(64, (3,3), activation='relu'),

# Second Max Pooling Layer
# Again reduces the feature map size
MaxPooling2D((2,2)),

# Flatten Layer
# Converts the 2D feature maps into a 1D vector
# so it can be connected to Dense (Fully Connected) layers
Flatten(),

# Fully Connected (Dense) Layer
# 128 neurons
# Learns high-level patterns for classification
Dense(128, activation='relu'),

# Output Layer
# 2 neurons because there are 2 classes
# Softmax converts outputs into probabilities
Dense(2, activation='softmax')

])

# Display the architecture of the CNN
model.summary()
```

```python
#Complie and Predict
model.compile(
optimizer="adam",
loss="categorical_crossentropy",
metrics=["accuracy"]
)
prediction = model.predict (image_array)
print (prediction)

predicted_class = np.argmax (prediction)
# e.g. 0 -> Cat, 1 -> Dog

print (predicted_class)
```

```python
#21,24


# Build CNN model
cnn_model = Sequential([

    # First Convolution Block
    Conv2D(
        32,
        (3, 3),
        activation="relu",
        input_shape=(32, 32, 3)
    ),

    BatchNormalization(),

    MaxPooling2D((2, 2)),

    # Second Convolution Block
    Conv2D(
        64,
        (3, 3),
        activation="relu"
    ),

    BatchNormalization(),

    MaxPooling2D((2, 2)),

    # Fully Connected Layers
    Flatten(),

    Dense(
        128,
        activation="relu"
    ),

    Dropout(0.5),

    Dense(
        10,
        activation="softmax"
    )
])

#22
# Compile model
cnn_model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

cnn_model.summary()
```

```python

import numpy as np
from tensorflow.keras.utils import to_categorical

# Create 100 random images of size 32x32x3
x_train = np.random.rand(100, 32, 32, 3)

# Create 100 random labels (0-9)
y_train = np.random.randint(0, 10, 100)
y_train = to_categorical(y_train, 10)

# Create validation data
x_validation = np.random.rand(20, 32, 32, 3)

y_validation = np.random.randint(0, 10, 20)
y_validation = to_categorical(y_validation, 10)
```

```python
#25
from tensorflow.keras.callbacks import EarlyStopping

# Create EarlyStopping callback
early_stopping = EarlyStopping(
    monitor="val_loss",
    patience=3,
    restore_best_weights=True
)

# Train the model
training_history = cnn_model.fit(
    x_train,
    y_train,
    epochs=15,
    batch_size=32,
    validation_data=(x_validation, y_validation),
    callbacks=[early_stopping]
)
```
