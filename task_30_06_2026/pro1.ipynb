```python
try:
    review = input("enter a product review: ")

    if review.strip() == "":
        raise ValueError("review cannot be empty")

    if review.isdigit():
        raise ValueError("review should contain only text, no number")

    review = review.lower()

    positivewords = ["good", "great", "excellent", "amazing"]
    negativewords = ["bad", "poor", "worst"]

    positive_count = sum(word in review for word in positivewords)
    negative_count = sum(word in review for word in negativewords)

    if positive_count > negative_count:
        print("Prediction: Positive 😊")
    elif negative_count > positive_count:
        print("Prediction: Negative ☹️")
    else:
        print("Prediction: Neutral")

except ValueError as e:
    print("input error:", e)

except Exception as e:
    print("unexpected error:", e)

else:
    print("review analysed successfully")

finally:
    print("AI prediction completed")
```
