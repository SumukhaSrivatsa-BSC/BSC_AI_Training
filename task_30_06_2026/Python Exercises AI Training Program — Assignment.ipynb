```python
# Ask the user to enter an message
message = input("Enter the message: ")

def detect_spam(message):
    message = message.lower()  # Convert to lowercase for case-insensitive checking
    
    # List of spam keywords
    spam_keywords = ["congratulations", "prize", "winner", "free", "offer"]
    
    # Check if any spam keyword is in the message
    for word in spam_keywords:
        if word in message:
            return "Spam"
    
    return "Not Spam"

detected_message = detect_spam(message)
print(f"The message is: {detected_message}")
```

```python
def calculate_accuracy(correct, total):
    if total == 0:
        return "Cannot calculate accuracy (total is zero)"
    
    accuracy = (correct / total) * 100
    return f"{accuracy}%"

print(calculate_accuracy(45, 50))
```

```python
def count_spam_predictions(predictions):
    count = 0
    
    for prediction in predictions:
        if prediction == "Spam":
            count += 1
            
    return count

data = ["Spam", "Spam", "Not Spam"]
print(count_spam_predictions(data))
```

Q: Why do we use constructors? Explain using your own example.

Constructors are used to automatically initialize an object’s data when it is created. They help ensure that every object starts with the required values, without needing to set them manually each time.

Why we use constructors:
#To initialize attributes (like name, age, etc.) at the time of object creation
#To save time and avoid repetitive code
#To make sure objects are always created in a valid and ready-to-use state

Example (simple explanation):
Imagine a Student object. Every student must have a name and course

Without a constructor:
You would create the object first and then manually assign values.
You might forget to set some values.

With a constructor:
When you create the student, you must provide the name and course immediately.
This ensures the object is complete from the beginning.

Real-life analogy:
Think of a constructor like a registration form.
When you sign up, you must fill in your details (name, email, etc.).
Similarly, a constructor forces you to provide important details when creating an object.


Level 4 — The Four Pillars of OOP

```python
# Create the BankAccount class
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # Private variable

    # Method to deposit money
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: {amount}")
        else:
            print("Invalid deposit amount")

    # Method to withdraw money
    def withdraw(self, amount):
        if amount > self.__balance:
            print("Insufficient balance")
        elif amount > 0:
            self.__balance -= amount
            print(f"Withdrawn: {amount}")
        else:
            print("Invalid withdrawal amount")

    # Method to check balance
    def check_balance(self):
        print(f"Current Balance: {self.__balance}")


# Example usage
account = BankAccount(1000)

account.deposit(500)
account.withdraw(200)
account.check_balance()
```

```python
# Parent class
class AIModel:
    def __init__(self, model_name):
        self.model_name = model_name

    def show_model(self):
        print(f"Model Name: {self.model_name}")


# Child class 1
class SpamDetector(AIModel):
    def detect(self):
        print("Spam Detection in progress...")


# Child class 2
class FaceRecognition(AIModel):
    def recognize(self):
        print("Face Recognition in progress...")


# Example usage
model1 = SpamDetector("Spam Detector")
model2 = FaceRecognition("Face Recognition")

model1.show_model()
model1.detect()

print()

model2.show_model()
model2.recognize()
```

```python
# Base class
class AIModel:
    def predict(self):
        print("Generic prediction")


# Child class 1
class SpamDetector(AIModel):
    def predict(self):
        print("SpamDetector: Checking if the message is spam")


# Child class 2
class FaceRecognition(AIModel):
    def predict(self):
        print("FaceRecognition: Identifying the person in the image")


# Child class 3
class RecommendationSystem(AIModel):
    def predict(self):
        print("RecommendationSystem: Suggesting items to the user")


# Example usage
models = [
    SpamDetector(),
    FaceRecognition(),
    RecommendationSystem()
]

# Polymorphism in action
for model in models:
    model.predict()
```

```python
from abc import ABC, abstractmethod

# Abstract class
class Model(ABC):

    @abstractmethod
    def train(self):
        pass


# Child class 1
class SpamDetector(Model):
    def train(self):
        print("Training the Spam Detector model")


# Child class 2
class ImageClassifier(Model):
    def train(self):
        print("Training the Image Classifier model")


# Example usage
model1 = SpamDetector()
model2 = ImageClassifier()

model1.train()
model2.train()
```

Q: Which OOP pillar is used when we hide data from users? Explain with one real-world example. 

The OOP pillar used when we hide data from users is called Encapsulation.

Explanation:
Encapsulation means keeping data (variables) private inside a class and only allowing access through specific methods. This prevents direct access and protects the data from misuse or accidental changes.

Real-world example:
Think of a bank account,
Your account balance is hidden (private).
You cannot directly change the balance.
Instead, you use methods like:
    Deposit
    Withdraw
    Check balance
This ensures:
You cannot accidentally set a negative balance.
All transactions follow proper rules.

```python
import logging
from datetime import datetime

# Configure logging
logging.basicConfig(
    filename="ai_system.log",
    level=logging.INFO,
    format="%(asctime)s - %(message)s",
    force=True
)
# SpamDetector Class
class SpamDetector:
    def __init__(self, model_name, version):
        self.model_name = model_name
        self.version = version

    # Predict method
    def predict(self, message):
        if not message.strip():
            raise ValueError("Message cannot be empty!")

        # Simple spam keywords logic
        spam_keywords = ["congratulations", "prize", "winner", "free", "offer"]
        message_lower = message.lower()

        for word in spam_keywords:
            if word in message_lower:
                return "Spam", 94   # confidence example

        return "Not Spam", 85   # confidence example


# Function to save prediction
def save_prediction(result):
    with open("prediction.txt", "a") as file:
        file.write(f"{result}\n")


# Main program
try:
    # Create object
    detector = SpamDetector("Spam Detector", "v1.0")

    # User input
    message = input("Enter your message: ")

    # Prediction
    prediction, confidence = detector.predict(message)

    # Display result
    print(f"Prediction: {prediction}")
    print(f"Confidence: {confidence}%")

    # Save to file
    save_prediction(f"{prediction} - {confidence}%")

    # Log the result
    log_msg = f"Prediction : {prediction} Confidence : {confidence}%"
    logging.info(log_msg)

except ValueError as e:
    print(f"Error: {e}")
```

Q35. OOP + File Handling 
A Student class was created, but every time the program ends, all student information is lost. Which Python concept would you use to make the data available even after the program closes? Explain your approach and write the code.

ANS: To keep data even after the program ends, we use File Handling concept in python.

Explanation:
Normally, data stored in variables or objects is lost when the program stops.
To save it permanently, we write the data to a file.
Later, we can read the file and restore the data.
This process is called data persistence.

Approach
Create a Student class
Save student details into a file (e.g. students.txt)
Data is written to a file using "a" (append mode)
File keeps data even after program ends 
When the program runs again, old data is still there
Read the file when needed

```python
# Student class
class Student:
    def __init__(self, name, age, course):
        self.name = name
        self.age = age
        self.course = course

    # Save student data to file
    def save_to_file(self):
        with open("students.txt", "a") as file:
            file.write(f"{self.name}, {self.age}, {self.course}\n")

    # Display student data
    def display(self):
        print(f"Name: {self.name}, Age: {self.age}, Course: {self.course}")


# Create student object
s1 = Student("Rahul", 20, "AI")

# Save data
s1.save_to_file()

print("Student data saved.\n")

# Read and display all students
print("Stored Students Records:")
with open("students.txt", "r") as file:
    content = file.read()
    print(content)
```
