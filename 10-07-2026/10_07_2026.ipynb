````python
import numpy as np

X = np.array([[1],[2]])
Y = np.array([[2], [4]])

weight = 0.5
bias = 0.1
learning_rate = 0.1

for epoch in range(10):
  Y_pred = X * weight + bias
  loss = np.mean((Y - Y_pred) ** 2)

  #Gradient Descent(GD) optimization
  dw =(-2/len(X)) * np.sum(X *(Y - Y_pred))#Derivative of weight
  db =(-2/len(X)) * np.sum(Y - Y_pred)#Derivative of bias

  weight = weight - learning_rate * dw
  bias = bias - learning_rate * db

  print(f"Epoch {epoch+1}")
  print("Loss:", round(loss,4))
````

````python
import numpy as np

X = np.array([1, 2, 3, 4])
Y = np.array([2, 4, 6, 8])

weight =0.5
bias = 0.1
learning_rate =0.1

for x, y in zip(X, Y):
  y_pred = weight * x + bias
  error = y - y_pred

  dw = -2 * x * error
  db = -2 * error

  weight = weight - learning_rate * dw
  bias = bias - learning_rate * db

  print(f"Input = {x}, Actual={y}")
  print(f"Prediction = {round(y_pred,2)}")
  print(f"Updated Weight = {round(weight,2)}")

print("Final Weight:",round(weight,2))
````

````python
#Adpactive Moment Estimation(Adam)
import numpy as np
X=np.array([[1],[2]])
Y=np.array([[2],[4]])
````

````python
weight=np.array([[0.5]])
bias=np.array([[0.1]])
````

````python
learning_rate = 0.1

epochs = 100 # Increased epochs for Adam to converge better

# Adam Optimizer hyperparameters
beta1 = 0.9
beta2 = 0.999
epsilon = 1e-8

# Initialize Adam momentum and RMSprop terms
m_weight = np.zeros_like(weight)
v_weight = np.zeros_like(weight)
m_bias = np.zeros_like(bias)
v_bias = np.zeros_like(bias)

for epoch in range(epochs):
  # Step 1: Forward Propagation (calculate predictions with current weights)
  Y_pred_current = np.dot(X, weight) + bias

  # Calculate the loss with current predictions (before update)
  loss_current = np.mean((Y - Y_pred_current) ** 2)

  # Calculate Gradients (dw, db) for Gradient Descent
  # These are crucial for the update rules and are derived from the MSE loss.
  dw = (-2 / len(X)) * np.sum(X * (Y - Y_pred_current))
  db = (-2 / len(X)) * np.sum(Y - Y_pred_current)

  # Adam Optimizer updates
  # Update biased first moment estimate (momentum)
  m_weight = beta1 * m_weight + (1 - beta1) * dw
  m_bias = beta1 * m_bias + (1 - beta1) * db

  # Update biased second moment estimate (RMSprop)
  v_weight = beta2 * v_weight + (1 - beta2) * (dw ** 2)
  v_bias = beta2 * v_bias + (1 - beta2) * (db ** 2)

  # Compute bias-corrected first moment estimate
  m_weight_corrected = m_weight / (1 - beta1 ** (epoch + 1))
  m_bias_corrected = m_bias / (1 - beta1 ** (epoch + 1))

  # Compute bias-corrected second moment estimate
  v_weight_corrected = v_weight / (1 - beta2 ** (epoch + 1))
  v_bias_corrected = v_bias / (1 - beta2 ** (epoch + 1))

  # Adam update rule for weight and bias
  weight = weight - learning_rate * m_weight_corrected / (np.sqrt(v_weight_corrected) + epsilon)
  bias = bias - learning_rate * m_bias_corrected / (np.sqrt(v_bias_corrected) + epsilon)

  # Step 8: Forward Propagation Again (calculate predictions with NEWLY updated weights)
  Y_pred_updated = np.dot(X, weight) + bias

  # Step 9: New Loss (calculate loss with NEWLY updated predictions)
  loss_updated = np.mean((Y - Y_pred_updated) ** 2)

  print(f"Epoch {epoch+1}:")
  print(f"  Loss before update: {round(loss_current, 4)}")
  print(f"  Loss after update: {round(loss_updated, 4)}")
  print(f"  Updated Weight: {round(weight[0][0], 4)}, Updated Bias: {round(bias[0][0], 4)}")
  print("-" * 30)

print(f"Final Weight: {round(weight[0][0], 4)}")
print(f"Final Bias: {round(bias[0][0], 4)}")
````

````python
#Activation Fuctions
#Binary step function
import numpy as np
import matplotlib.pyplot as plt

# Binary Step Function
def binary_step (x):
    return np.where (x >= 0, 1, 0)

# Generate Input Values
x = np.linspace (-5, 5, 100) #line space (start,end,divide into100 parts )

# Apply Function
y = binary_step(x)

print("Input :", x [: 10] )
print("Output:", y[ : 10])

plt.plot(x, y, linewidth=3)
plt.title("Binary Step Activation")
plt.grid(True)
plt.show()
````

->During training, all neurons receive identical updates.

->Neurons learn the same features instead of different features.

->The network cannot effectively learn complex patterns.

->This problem is called symmetry. Therefore, weights are usually initialized with small random values rather than zeros.
